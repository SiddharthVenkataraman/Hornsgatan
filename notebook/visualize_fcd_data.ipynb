```python
import pandas as pd
import folium
import os
import pandas as pd

project_root = os.environ["HORNSGATAN_HOME"]
os.chdir(project_root)
print("Project root is:", project_root)

date = "2020-01-02"
vehicles_number = 1 
fcd_from_calibration = True


if vehicles_number==0:
    prefix = f"{date}"
else:
    prefix = f"{date}_{vehicles_number}"

if fcd_from_calibration:
    path= "data/calibration_data/"
    csv_file_path_e2w_out = f'{path}fcd_data_e2w_out_{prefix}.csv'
    csv_file_path_e2w_in = f'{path}fcd_data_e2w_in_{prefix}.csv'
    csv_file_path_w2e_out = f'{path}fcd_data_w2e_out_{prefix}.csv'
    csv_file_path_w2e_in = f'{path}fcd_data_w2e_in_{prefix}.csv'
else:
    path = "data/sim_data/"
    csv_file_path_e2w_out = f'{path}fcd_output_e2w_out_{prefix}.csv'
    csv_file_path_e2w_in = f'{path}fcd_output_e2w_in_{prefix}.csv'
    csv_file_path_w2e_out = f'{path}fcd_output_w2e_out_{prefix}.csv'
    csv_file_path_w2e_in = f'{path}fcd_output_w2e_in_{prefix}.csv'

# Read the CSV file  IN into a pandas DataFrame
df_e2w_in = pd.read_csv(csv_file_path_e2w_in)
#df_e2w_in["id"] = df_e2w_in["id"].apply(lambda x: int(x.split('_')[0]))
#df_e2w_in = df_e2w_in[df_e2w_in["id"]==5 ]
# Read the CSV file OUT into a pandas DataFrame
df_e2w_out = pd.read_csv(csv_file_path_e2w_out)
#df_e2w_out["id"] = df_e2w_out["id"].apply(lambda x: int(x.split('_')[0]))
#df_e2w_out = df_e2w_in[df_e2w_in["id"]==5 ]

# Read the CSV file  IN into a pandas DataFrame
df_w2e_in = pd.read_csv(csv_file_path_w2e_in)
#df_w2e_in["id"] = df_w2e_in["id"].apply(lambda x: int(x.split('_')[0]))

# Read the CSV file OUT into a pandas DataFrame
df_w2e_out = pd.read_csv(csv_file_path_w2e_out)
#df_w2e_out["id"] = df_w2e_out["id"].apply(lambda x: int(x.split('_')[0]))

    
```

```python
df_e2w_in.head()
```

```python

mean_lat = df_e2w_in['y'].mean()
mean_lon = df_e2w_in['x'].mean()
m = folium.Map(location=[mean_lat, mean_lon], zoom_start=12)

# Add circle markers for each data point
# It's good practice to handle potential NaNs in location columns
for index, row in df_e2w_in.iterrows():
    # Check for NaN values in latitude and longitude
        folium.CircleMarker(
            [row['y'], row['x']],
            radius=5, # Define the size of the circle marker in pixels
            color='blue', # Border color
            fill=True,
            fill_color='blue', # Fill color
            fill_opacity=0.6, # Opacity of the fill color
            popup=f"id = {row["id"]}, speed = {row["speed"]}, lane={row['lane']}, time = {row['time']}",
            tooltip=f"id = {row["id"]}, speed = {row["speed"]}, lane = {row['lane']}, time = {row['time']}",

        
        ).add_to(m)
        
for index, row in df_e2w_out.iterrows():
    # Check for NaN values in latitude and longitude
        folium.CircleMarker(
            [row['y'], row['x']],
            radius=5, # Define the size of the circle marker in pixels
            color='red', # Border color
            fill=True,
            fill_color='red', # Fill color
            fill_opacity=0.6, # Opacity of the fill color
            popup=f"id = {row["id"]}, speed = {row["speed"]}, lane={row['lane']}, time = {row['time']}",
            tooltip=f"id = {row["id"]}, speed = {row["speed"]}, lane = {row['lane']}, time = {row['time']}",

        ).add_to(m)

for index, row in df_w2e_in.iterrows():
    # Check for NaN values in latitude and longitude
        folium.CircleMarker(
            [row['y'], row['x']],
            radius=5, # Define the size of the circle marker in pixels
            color='blue', # Border color
            fill=True,
            fill_color='blue', # Fill color
            fill_opacity=0.6, # Opacity of the fill color
            popup=f"id = {row["id"]}, speed = {row["speed"]}, lane={row['lane']}, time = {row['time']}",
            tooltip=f"id = {row["id"]}, speed = {row["speed"]}, lane = {row['lane']}, time = {row['time']}",

        ).add_to(m)
        
for index, row in df_w2e_out.iterrows():
    # Check for NaN values in latitude and longitude
        folium.CircleMarker(
            [row['y'], row['x']],
            radius=5, # Define the size of the circle marker in pixels
            color='red', # Border color
            fill=True,
            fill_color='red', # Fill color
            fill_opacity=0.6, # Opacity of the fill color
            popup=f"id = {row["id"]}, speed = {row["speed"]}, lane={row['lane']}, time = {row['time']}",
            tooltip=f"id = {row["id"]}, speed = {row["speed"]}, lane = {row['lane']}, time = {row['time']}",
   ).add_to(m)
    
# Display the map (or save to an HTML file)
# To display in a Jupyter notebook/Lab, the map object itself is usually enough
# To save to HTML, uncomment the line below:
# m.save('fcd_data_map.html')
print("Map created. Displaying map object...")

  
```
