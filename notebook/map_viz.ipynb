```python
import sumolib
import pandas as pd
from shapely.geometry import LineString, Polygon


# Load your SUMO network
net = sumolib.net.readNet("Hornsgatan_v2.net.xml")
edge_lane_data = []

def getShape(lane):
    _shape = list()
    for point in lane.getShape():
        _shape.append(net.convertXY2LonLat(point[0], point[1]))
    return LineString(_shape)

# Prepare a list to hold edge-lane info
for edge in net.getEdges():
    for lane in edge.getLanes():
        edge_lane_data.append({
            "edge_id": edge.getID(),
            "lane_id": lane.getID(),
            "lane_index": lane.getIndex(),
            "length": lane.getLength(),
            "speed": lane.getSpeed(),
            "from_node": edge.getFromNode().getID(),
            "to_node": edge.getToNode().getID(),
            "num_lanes": edge.getLaneNumber(),
            "edge_name": edge.getName(),
            "edge_type": edge.getType(),
            "lane_shape_lonlat": getShape(lane)
            # Uncomment below to include lon/lat as well:
            # "lane_shape_lonlat": ";".join([f"{lon},{lat}" for lon, lat in shape_lonlat])
        })

df = pd.DataFrame(edge_lane_data)
df.to_csv("all_edge_lanes_with_shape.csv", index=False)
```

```python
def getInteralEdgeid(fromedgeid , toedgeid, net):
    fromedge = net.getEdge(fromedgeid)
    toedge = net.getEdge(toedgeid)
    conlist = fromedge.getOutgoing()[toedge]
    return net.getLane(conlist[0].getViaLaneID()).getEdge().getID()
```

```python
def createRouteInternal(net, route):
    mylist = list()
    route_list = route.split(" ")
    for i in range(len(route_list)-1):
        mylist.append(route_list[i])
        mylist.append(getInteralEdgeid(route_list[i] , route_list[i+1], net))
    mylist.append(route_list[-1])
    return " ".join(mylist)
```

```python
def getinoutedgesfrominternal(internalid):
    internaledge = net.getEdge(internalid)
    fromid = list(internaledge.getIncoming().keys())[0].getID()
    toid = list(internaledge.getOutgoing().keys())[0].getID()
    return fromid, toid
```

```python
from shapely.geometry import LineString, Polygon

def getShape(edge):
    _shape = list()
    for point in edge.getShape():
        _shape.append(net.convertXY2LonLat(point[0], point[1]))
    return LineString(_shape)
```

```python
edge_data = list()
internal_edge_data = list()
Hornsgatan_only = list()
for edge in net.getEdges():
    sh = edge.getShape()
    #first = edge.getFromNode().getShape()
    #print(first)
    #end = edge.getToNode().getShape()
    #sh = [first] + sh + [end]
    if len(sh) > 0:
        if sh[0]==sh[-1]:
            sh.pop()
        if len(sh)>1:
            shape = list()
            for point in sh:
                shape.append(net.convertXY2LonLat(point[0], point[1]))
            shape = LineString(shape) 
            #dilated = shape.buffer(edge.getLaneNumber()*0.000001)
            edge_data.append({"edge_id":str(edge.getID()), "shape" :shape})
            if edge.getName() == "Hornsgatan":
                Hornsgatan_only.append({"edge_id":str(edge.getID()), "shape" :shape})
```
